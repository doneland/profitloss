```python
import pandas as pd
from datetime import datetime

from price import Price
from strategy import Strategy

eurusd_df = pd.read_csv('EURUSD.csv')
eurusd_df.shape
```

```python
eurusd_df.head(3)
```

```python
dates = eurusd_df['Date']
dates = pd.to_datetime(dates, format='%d/%m/%Y %H:%M')
eurusd_df['Date'] = dates
sorted_prices = eurusd_df.sort_values(by=['Date'], ascending=[-1,])
```

```python
columns = ['Date', 'Close']
prices = sorted_prices[columns]
```

```python
s = Strategy(start_amount=1000, sl=0.001, tpmulti=3, leverage=100)
for p in prices.values:
    p = Price(p)
    s.stream_price(p)

s.stats
```

```python
s.status
```
